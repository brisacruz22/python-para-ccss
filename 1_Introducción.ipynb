### ¿Cómo sé la codificación de la base de datos que estoy usando?

La mayoría es UTF-8 o ISO-8859-10. Usaremos la función chardet.detect

```python
!pip install cchardet
import cchardet as chardet
```

```python
# Detectar la codificación del archivo
def detect_encoding(file_path):
    with open(file_path, 'rb') as file:
        return chardet.detect(file.read())['encoding']

# Usar la función
encoding = detect_encoding('Datos/Enaho01A-2023-300.csv')
print(encoding)

```

```python
# Detectar la codificación del archivo
def detect_encoding(file_path):
    with open(file_path, 'rb') as file:
        return chardet.detect(file.read())['encoding']

# Usar la función
encoding = detect_encoding('Datos/Latinobarometro_2023_Esp_Spss.sav')
print(encoding)
```

.size: Devuelve el número total de elementos en el DataFrame.

### Tipos de Datos en Pandas
En pandas, cada columna de un DataFrame puede tener un tipo de dato específico, conocido como dtype. Los dtypes más comunes son:

- object: Para textos o combinaciones de texto y números.
- int64: Para números enteros.
- float64: Para números con decimales.
- bool: Para valores verdadero o falso (True o False).
- datetime64: Para fechas y horas.
- category: Para datos categóricos con un número limitado de categorías.

Estos tipos de datos ayudan a pandas a manejar de manera eficiente diferentes tipos de información, asegurando que las operaciones sobre los datos sean rápidas y que el uso de memoria sea óptimo.

### Tipos de Datos en Pandas
En pandas, cada columna de un DataFrame puede tener un tipo de dato específico, conocido como dtype. Los dtypes más comunes son:

- object: Para textos o combinaciones de texto y números.
- int64: Para números enteros.
- float64: Para números con decimales.
- bool: Para valores verdadero o falso (True o False).
- datetime64: Para fechas y horas.
- category: Para datos categóricos con un número limitado de categorías.

Estos tipos de datos ayudan a pandas a manejar de manera eficiente diferentes tipos de información, asegurando que las operaciones sobre los datos sean rápidas y que el uso de memoria sea óptimo.
